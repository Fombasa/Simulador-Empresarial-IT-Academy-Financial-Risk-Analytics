# Data Cleaning Informe ECF 2021

## Librerias

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

## Carga del fichero csv

```python
df = pd.read_csv(
    r"..\Data\Bronze_ECF_090226.csv",
    sep=","
)
df.shape
```

```python
df.info()
```

```python
df.head()
```

## 1. Tratamiento de valores NULL 

```python
columns_with_null = df.columns[df.isnull().any()].tolist()
columns_with_null
```

```python
# 1) Eliminación de nulls con una ocurrencia inferior al 3% dentro de la variable en cuestión (household_type, residence_type, deuda, adq_deuda2)
df = df.dropna(subset=["familia", "riq_inmob","deuda", "adq_deuda2"])
```

```python
columns_with_null = df.columns[df.isnull().any()].tolist()
columns_with_null
```

```python
# 2) Cambio de nulls con una ocurrencia superior al 3% por -1 (inflacion_ok, interes_comp_ok, diversificacion_ok) 

df[["inflacion_ok", "interes_comp_ok", "diversificacion_ok"]] = (
    df[["inflacion_ok", "interes_comp_ok", "diversificacion_ok"]]
    .fillna(-1)
)
```

```python
columns_with_null = df.columns[df.isnull().any()].tolist()
columns_with_null
```

## 2. Estandarización de formatos

```python
df.info()
```

```python
# Cambio de literal int a variables binarias
df["ahorro"] = df["ahorro"].astype(int)
df["deuda"] = df["deuda"].astype(int)
df["adq_ahorro2"] = df["adq_ahorro2"].astype(int)
df["adq_deuda2"] = df["adq_deuda2"].astype(int)
df["adq_otro"] = df["adq_otro"].astype(int)
df["inflacion_ok"] = df["inflacion_ok"].astype(int)
df["interes_comp_ok"] = df["interes_comp_ok"].astype(int)
df["diversificacion_ok"] = df["diversificacion_ok"].astype(int)
```

## 3. Validación final 

```python
# NULL summary 
null_counts = df.isna().sum()
null_summary = (
    null_counts[null_counts > 0]
    .sort_values(ascending=False)
    .rename("null_count")
    .reset_index()
    .rename(columns={"index": "column"})
)

print("=== NULLS (only columns with >0) ===")
display(null_summary)
print()
```

```python
# Types
types_summary = (
    df.dtypes.astype(str)
    .rename("dtype")
    .reset_index()
    .rename(columns={"index": "column"})
)

print("=== DTYPES ===")
display(types_summary)
print()
```

## Exportar el dataset limpio (Silver)

```python
df.to_csv(
    r"..\Data\Silver_ECF_090226.csv",
    index=False,
    encoding="utf-8"
)
```
