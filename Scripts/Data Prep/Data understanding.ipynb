### Data understanding 

```python
total_registros = df["id"].count()
total_registros
```

```python
ids_repetidos = (
    df
    .groupby("id")
    .size()
    .reset_index(name="repeated")
    .query("repeated > 1")
)

ids_repetidos
```

```python
total_ids_unicos = df["id"].nunique()
total_ids_unicos
```

```python
columnas = [
    "id", "age", "job", "marital", "education", "balance",
    "housing", "loan", "contact", "day", "month", "duration",
    "campaign", "pdays", "previous", "poutcome", "deposit"
]

duplicados_absolutos = (
    df
    .groupby(columnas)
    .size()
    .reset_index(name="cuenta")
    .query("cuenta > 1")
)

duplicados_absolutos
```
